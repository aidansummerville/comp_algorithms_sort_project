```python
## taken from https://www.geeksforgeeks.org/merge-sort/


##Merge sort Function
def mergeSort(arr): 
    
    ## if the length of the array is greater than 1
    if len(arr) >1: 
        ##finds the middle point of the array
        mid = len(arr)//2 
        ## divies array into a left half
        L = arr[:mid] 
        ##divies array into a right half
        R = arr[mid:] 
  
        mergeSort(L) # calls mergesort again for left half 
        mergeSort(R) # calls mergesort again for right half
  ## repeats until all elemnts are individual

      ## i, j and k set to 0
        i = j = k = 0
          
    ##copys datat o tempory arrays and sorts them to accending order 
        while i < len(L) and j < len(R): 
            if L[i] < R[j]: 
                arr[k] = L[i] 
                i+=1
            else: 
                arr[k] = R[j] 
                j+=1
            k+=1
          
        # Checks for left over elements in the left and right arrays
        while i < len(L): 
            arr[k] = L[i] 
            i+=1
            k+=1
          
        while j < len(R): 
            arr[k] = R[j] 
            j+=1
            k+=1
  ## all repeats until array comes back together sorted
```

```python

## code taken from https://www.geeksforgeeks.org/insertion-sort/
## own comments added.


##Insertion sort Funtion
def insertionSort(arr): 
  
   ##Starts at the 2nd position within the array and iterates through until last position withon the array 
    for i in range(1, len(arr)): 
        
        
        ##Selects the integer within the array at point i intially the 2nd position and sets it as the key
        key = arr[i] 

        
        ##sets j to the position before the key 
        j = i-1
        ##while j is greater than or = to 0  and key is less than j
        while j >= 0 and key < arr[j] : 
               ## j moves to next position in the array if key is less than j
                arr[j + 1] = arr[j] 
                j -= 1
        ## key is now in the next position in the array
        arr[j + 1] = key 
```

```python
##code taken from https://simply-python.com/2018/11/15/counting-sort-in-python/
k = 50001 ## k is the range of the possible numbers +1
## count sort function
def countSort(arr, k): 
    
 
    # creates index set from 0 to k the max possible value of the array
    count_list = [0]*(k)
 
    # loop through the arr and add 1 for every occurence oput into count list
    for n in arr:
        count_list[n] = count_list[n] + 1
 
    # Sorts arrays back into the original list using the position of occurecne in the array.
    i=0
    ## in the range of the count list
    for n in range(len(count_list)):
        ## if n is greater than 0
        while count_list[n] > 0:
            ## sort each interger into its positoon in the array and minus the count by 1 count_list in that position everytime
            arr[i] = n
            i+=1
            count_list[n] -= 1
```

```python
## code taken from https://www.geeksforgeeks.org/bubble-sort/

## calls bubblesort function
def bubbleSort(arr):
    ## n = length of the array
    n = len(arr)
 
    ##iterate through all the elements in the array
    for i in range(n):
 
        ##iterate through list in range length array -1 and i
        ## list gets smaller everytime first for loop completes
        for j in range(0, n-i-1):
 
            ##swap the element j with the elemet beside it if it is greater than it.
            if arr[j] > arr[j+1] :
                arr[j], arr[j+1] = arr[j+1], arr[j]

```

```python
## http://interactivepython.org/lpomz/courselib/static/pythonds/SortSearch/TheSelectionSort.html


## call selection sort
def selectionSort(arr):
    ## sets fill slot to length of array -1
    ## fillslot set to range array -1 then -1 verytime the loop completes until 0
   for fillslot in range(len(arr)-1,0,-1):
    ## postion of max = 0
       positionOfMax=0
        ## location is 1 above fillslot
       for location in range(1,fillslot+1):
        ## if location is greater than max location is max
           if arr[location]>arr[positionOfMax]:
               positionOfMax = location
      ## holds tempory postion of number swapped for max
       temp = arr[fillslot]
       arr[fillslot] = arr[positionOfMax]
       arr[positionOfMax] = temp
```

```python
## creates random array of size n in range 0 to 50000

def random_array(n):
  array =[]
  for i in range( 0, n, 1):
    array.append(randint(0, 50000))
  return array 
```

```python
## code for creating  average times of completion across each n input for each sorting algorithm

##  for i in each of the input sizes
for i in range(0,len(size)):
  timesi = []  ##creates arrays for each algorithm that the length of completion for that input size will be added to before an avaereg is taken
  timesm = []
  timesc = []
  timesb = []
  timess = []
  for j in range(9):## runs each input size 10 times
   if __name__ == '__main__': ## calls the main function
    arr = random_array(size[i])## calls the random array function for the input size
    print(arr)  ## prints unsorted array(not nessasry but is usefull to veiwing purposes)
    ##mergesort timer
    startm = time.time()  ## starts timer
    mergeSort(arr) ## calls mergesort
    endm = time.time()## ends time
    print(arr)  ## prints sorted array(not nessasry but is usefull to veiwing purposes)
    time_elapsedm = endm - startm  ## minus start time from end time to get completion time
    timesm.append(time_elapsedm)  ##appends completion time to times m array
    ## insertion timer
    arr = random_array(size[i])
    print(arr)
    starti = time.time()    ## below algorithm timers work in the same way
    insertionSort(arr)
    endi = time.time()
    print(arr)
    time_elapsedi = endi - starti
    timesi.append(time_elapsedi)
    ## countsort timer
    arr = random_array(size[i])
    print(arr)
    startc = time.time()  
    countSort(arr,k) 
    endc = time.time()
    print(arr)
    time_elapsedc = endc - startc
    timesc.append(time_elapsedc)
    ##bublesorttimer
    arr = random_array(size[i])
    print(arr)
    startb = time.time()  
    bubbleSort(arr)
    endb = time.time()
    print(arr)
    time_elapsedb = endb - startb
    timesb.append(time_elapsedb)
    ##selectiontimer
    arr = random_array(size[i])
    print(arr)
    starts = time.time() 
    selectionSort(arr)
    ends = time.time()
    print(arr)
    time_elapseds = ends - starts
    timess.append(time_elapseds)
  if i == 0: ## for the iput size of n = 100 get an average for each algorthim and apend to t100
   t100.append((sum(timesm)/len(timesm))*1000)
   t100.append((sum(timesi)/len(timesi))*1000)
   t100.append((sum(timesc)/len(timesc))*1000)
   t100.append((sum(timesb)/len(timesb))*1000) ##dived by 1000 for miliseconds
   t100.append((sum(timess)/len(timess))*1000)
  if i == 1: ## for the iput size of n = 350 get an average for each algorthim and apend to t250
   t250.append((sum(timesm)/len(timesm))*1000)
   t250.append((sum(timesi)/len(timesi))*1000)
   t250.append((sum(timesc)/len(timesc))*1000) ##dived by 1000 for miliseconds
   t250.append((sum(timesb)/len(timesb))*1000)
   t250.append((sum(timess)/len(timess))*1000)  
  if i == 2: ## for the iput size of n = 500 get an average for each algorthim and apend to t500
   t500.append((sum(timesm)/len(timesm))*1000)
   t500.append((sum(timesi)/len(timesi))*1000)
   t500.append((sum(timesc)/len(timesc))*1000) ##dived by 1000 for miliseconds
   t500.append((sum(timesb)/len(timesb))*1000)
   t500.append((sum(timess)/len(timess))*1000)
  if i == 3: ## for the iput size of n = 750 get an average for each algorthim and apend to t750
   t750.append((sum(timesm)/len(timesm))*1000)
   t750.append((sum(timesi)/len(timesi))*1000)
   t750.append((sum(timesc)/len(timesc))*1000)
   t750.append((sum(timesb)/len(timesb))*1000)
   t750.append((sum(timess)/len(timess))*1000)  
  if i == 4: ## for the iput size of n = 1000 get an average for each algorthim and apend to t1000
   t1000.append((sum(timesm)/len(timesm))*1000)
   t1000.append((sum(timesi)/len(timesi))*1000) ##dived by 1000 for miliseconds
   t1000.append((sum(timesc)/len(timesc))*1000)
   t1000.append((sum(timesb)/len(timesb))*1000)
   t1000.append((sum(timess)/len(timess))*1000)
  if i == 5: ## for the iput size of n = 1250 get an average for each algorthim and apend to t1250
   t1250.append((sum(timesm)/len(timesm))*1000)
   t1250.append((sum(timesi)/len(timesi))*1000)
   t1250.append((sum(timesc)/len(timesc))*1000) ##dived by 1000 for miliseconds
   t1250.append((sum(timesb)/len(timesb))*1000)
   t1250.append((sum(timess)/len(timess))*1000)  
  if i == 6: ## for the iput size of n = 2500 get an average for each algorthim and apend to t2500
   t2500.append((sum(timesm)/len(timesm))*1000)
   t2500.append((sum(timesi)/len(timesi))*1000)
   t2500.append((sum(timesc)/len(timesc))*1000) ##dived by 1000 for miliseconds
   t2500.append((sum(timesb)/len(timesb))*1000)
   t2500.append((sum(timess)/len(timess))*1000)
  if i == 7: ## for the iput size of n = 3750 get an average for each algorthim and apend to t3750
   t3750.append((sum(timesm)/len(timesm))*1000)
   t3750.append((sum(timesi)/len(timesi))*1000)
   t3750.append((sum(timesc)/len(timesc))*1000) ##dived by 1000 for miliseconds
   t3750.append((sum(timesb)/len(timesb))*1000)
   t3750.append((sum(timess)/len(timess))*1000)    
  if i == 8: ## for the iput size of n = 5000 get an average for each algorthim and apend to t5000
   t5000.append((sum(timesm)/len(timesm))*1000)
   t5000.append((sum(timesi)/len(timesi))*1000)
   t5000.append((sum(timesc)/len(timesc))*1000) ##dived by 1000 for miliseconds
   t5000.append((sum(timesb)/len(timesb))*1000)
   t5000.append((sum(timess)/len(timess))*1000)
  if i == 9: ## for the iput size of n = 6250 get an average for each algorthim and apend to t6250
   t6250.append((sum(timesm)/len(timesm))*1000)
   t6250.append((sum(timesi)/len(timesi))*1000)
   t6250.append((sum(timesc)/len(timesc))*1000) ##dived by 1000 for miliseconds
   t6250.append((sum(timesb)/len(timesb))*1000)
   t6250.append((sum(timess)/len(timess))*1000)  
  if i == 10: ## for the iput size of n = 7500 get an average for each algorthim and apend to t7500
   t7500.append((sum(timesm)/len(timesm))*1000)
   t7500.append((sum(timesi)/len(timesi))*1000)
   t7500.append((sum(timesc)/len(timesc))*1000) ##dived by 1000 for miliseconds
   t7500.append((sum(timesb)/len(timesb))*1000)
   t7500.append((sum(timess)/len(timess))*1000)
  if i == 11: ## for the iput size of n = 8750 get an average for each algorthim and apend to t8750
   t8750.append((sum(timesm)/len(timesm))*1000)
   t8750.append((sum(timesi)/len(timesi))*1000)
   t8750.append((sum(timesc)/len(timesc))*1000) ##dived by 1000 for miliseconds
   t8750.append((sum(timesb)/len(timesb))*1000)
   t8750.append((sum(timess)/len(timess))*1000)  
  if i == 12: ## for the iput size of n = 10000 get an average for each algorthim and apend to t10000
   t10000.append((sum(timesm)/len(timesm))*1000)
   t10000.append((sum(timesi)/len(timesi))*1000)
   t10000.append((sum(timesc)/len(timesc))*1000) ##dived by 1000 for miliseconds
   t10000.append((sum(timesb)/len(timesb))*1000)
   t10000.append((sum(timess)/len(timess))*1000)
```
